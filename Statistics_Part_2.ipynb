# 1 . What is hypothesis testing in statistics?

▶
**Hypothesis testing** is a statistical method used to decide whether there is enough evidence in a sample to support a specific claim about a population.

* **Null Hypothesis ($H_0$)**: No effect or no difference.
* **Alternative Hypothesis ($H_1$)**: There is an effect or a difference.
* You collect data, run a test (like a t-test), and calculate a **p-value**.
* If the p-value is small (usually < 0.05), you **reject $H_0$**.
* If it's large, you **fail to reject $H_0$**.


# 2 . What is the null hypothesis, and how does it differ from the alternative hypothesis?
▶
** Null Hypothesis ($H_0$)**:
It is the default or starting assumption in a hypothesis test. It states that there is **no effect**, **no difference**, or **no relationship** in the population.

> Example: $H_0: \mu = 50$ (The population mean is 50)

** Alternative Hypothesis ($H_1$ or $H_a$)**:
It is what you want to prove. It states that there **is** an effect, **a difference**, or **a relationship**.

> Example: $H_1: \mu \neq 50$ (The population mean is not 50)

---

###  Key Difference:

| Null Hypothesis ($H_0$)     | Alternative Hypothesis ($H_1$)    |
| --------------------------- | --------------------------------- |
| Assumes no change or effect | Assumes a change or effect        |
| Is tested directly          | Is supported if $H_0$ is rejected |
| Conservative assumption     | Reflects the researcher's claim   |

---

In short:
**$H_0$** = “nothing’s happening”
**$H_1$** = “something’s happening”


# 3 . What is the significance level in hypothesis testing, and why is it important?
▶

**Significance Level ($\alpha$)**:
It is the threshold probability used to decide whether to reject the null hypothesis.
Common values: 0.05, 0.01, 0.10

---

**Why It's Important:**

* It represents the risk of making a **Type I error** — rejecting the null hypothesis when it is actually true.
* It sets the standard for evidence: if the **p-value** is less than $\alpha$, the result is considered **statistically significant**, and you reject $H_0$.

---

**Example:**

If $\alpha = 0.05$:
You're accepting a 5% chance of wrongly rejecting a true null hypothesis.

---

**Summary:**

The **significance level** controls how strict your test is.

* Lower $\alpha$ = more cautious decision
* Higher $\alpha$ = more likely to detect an effect, but with more risk of error


# 4 . What does a P-value represent in hypothesis testing?
▶

**P-value** is the probability of obtaining test results at least as extreme as the observed results, **assuming the null hypothesis is true**.

---

**In simple terms:**
It tells you how likely your sample data is if the null hypothesis were actually correct.

---

**Interpretation:**

* **Low p-value (≤ α)**: Strong evidence against the null hypothesis → **Reject $H_0$**
* **High p-value (> α)**: Weak evidence against the null hypothesis → **Fail to reject $H_0$**

---

**Example:**
If you get a **p-value of 0.03** and your significance level is **0.05**, you reject the null hypothesis. There's only a **3% chance** you'd see data this extreme if $H_0$ were true.

---

**Summary:**
The **p-value** measures the strength of evidence **against** the null hypothesis. Smaller values mean stronger evidence.


# 5 . How do you interpret the P-value in hypothesis testing?
▶

The **p-value** helps you decide whether to reject the **null hypothesis**.

---

**Interpretation Guidelines:**

* **If p-value ≤ significance level ($\alpha$)**:

  * The result is **statistically significant**
  * **Reject** the null hypothesis ($H_0$)
  * There is evidence to support the alternative hypothesis ($H_1$)

* **If p-value > significance level ($\alpha$)**:

  * The result is **not statistically significant**
  * **Fail to reject** the null hypothesis
  * There is **not enough evidence** to support $H_1$

---

**Example (with $\alpha = 0.05$):**

* **p = 0.02** → 0.02 ≤ 0.05 → **Reject $H_0$** → Significant result
* **p = 0.10** → 0.10 > 0.05 → **Fail to reject $H_0$** → Not significant

---

**Summary:**

The **smaller the p-value**, the stronger the evidence **against** the null hypothesis.
Compare it to your chosen $\alpha$ (like 0.05) to make your decision.


# 6. What are Type 1 and Type 2 Errors in Hypothesis Testing?

In hypothesis testing, errors can happen because decisions are based on sample data, not the whole population.

---

**Type 1 Error (False Positive):**

* Occurs when you **reject the null hypothesis** ($H_0$) even though it is **true**.
* You think there is an effect or difference, but there isn't.
* The probability of making a Type 1 error is the **significance level $\alpha$** (commonly 0.05).

---

**Type 2 Error (False Negative):**

* Occurs when you **fail to reject the null hypothesis** ($H_0$) even though the **alternative hypothesis is true**.
* You miss detecting a real effect or difference.
* The probability of making a Type 2 error is denoted by $\beta$.

---

### Summary Table

| Error Type   | What Happens                    | Probability Symbol | Meaning        |
| ------------ | ------------------------------- | ------------------ | -------------- |
| Type 1 Error | Reject $H_0$ when $H_0$ true    | $\alpha$           | False positive |
| Type 2 Error | Fail to reject $H_0$ when false | $\beta$            | False negative |

---

**In short:**

* Type 1 = False alarm
* Type 2 = Missed detection


# 7. What is the Difference Between a One-Tailed and a Two-Tailed Test in Hypothesis Testing?

**One-Tailed Test:**

* Tests for an effect in **only one direction** (either greater than or less than).
* The alternative hypothesis specifies the direction.
* Example: $H_1: \mu > 50$ or $H_1: \mu < 50$
* The entire significance level ($\alpha$) is placed on one tail of the distribution.

---

**Two-Tailed Test:**

* Tests for an effect in **both directions** (whether the parameter is either greater or less than a certain value).
* The alternative hypothesis does **not** specify direction.
* Example: $H_1: \mu \neq 50$
* The significance level ($\alpha$) is split equally between both tails (e.g., 0.025 in each tail if $\alpha=0.05$).

---

### Summary:

| Aspect                 | One-Tailed Test                 | Two-Tailed Test                   |
| ---------------------- | ------------------------------- | --------------------------------- |
| Direction              | Tests one direction only        | Tests both directions             |
| Alternative Hypothesis | Directional ($>$ or $<$)        | Non-directional ($\neq$)          |
| Use when               | You expect a specific direction | You want to detect any difference |
| $\alpha$ distribution  | All in one tail                 | Split between two tails           |

---

In short:
Use a **one-tailed test** if you predict the effect's direction; otherwise, use a **two-tailed test**.


# 8. What is the Z-test, and When Is It Used in Hypothesis Testing?

**Z-test** is a statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means) when the population variance is known or the sample size is large.

---

### When to Use a Z-test:

* The population standard deviation ($\sigma$) is **known**.
* The sample size is **large** (usually $n > 30$)—then the sampling distribution of the sample mean is approximately normal by the Central Limit Theorem.
* You are testing hypotheses about population means or proportions.

---

### Common Types of Z-tests:

* **One-sample Z-test:** Test if a sample mean differs from a known population mean.
* **Two-sample Z-test:** Compare means from two independent samples (when population variances are known).
* **Z-test for proportions:** Test if a sample proportion differs from a known population proportion.

---

### Summary:

* Z-test uses the **standard normal distribution**.
* Appropriate when **population variance is known** or sample size is large.
* Often replaced by the **t-test** when population variance is unknown and sample size is small.


# 9. How Do You Calculate the Z-score, and What Does It Represent in Hypothesis Testing?

**Z-score** measures how many standard deviations a data point or sample mean is from the population mean under the null hypothesis.

---

### How to Calculate the Z-score:

For a sample mean:

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{x}$ = sample mean
* $\mu$ = population mean under the null hypothesis
* $\sigma$ = population standard deviation
* $n$ = sample size

---

### What the Z-score Represents:

* It tells you the position of the sample mean relative to the population mean in terms of standard deviations.
* A large absolute value of Z indicates that the sample mean is far from the null hypothesis mean, suggesting evidence against $H_0$.
* Used to find the **p-value** or compare to critical Z-values to decide whether to reject $H_0$.

---

### Summary:

The Z-score quantifies how unusual your sample result is if the null hypothesis is true.

* High $|Z|$ → less likely result under $H_0$ → possible rejection of $H_0$.


# 10. What is the T-distribution, and When Should It Be Used Instead of the Normal Distribution?

**T-distribution** (or Student’s t-distribution) is a probability distribution that is similar to the normal distribution but has heavier tails. This means it accounts for more variability in the data, especially with smaller samples.

---

### When to Use the T-distribution:

* The **sample size is small** (usually $n < 30$).
* The **population standard deviation ($\sigma$) is unknown**.
* You are estimating the population mean using the sample mean.
* The underlying data is approximately normally distributed.

---

### Why Use the T-distribution Instead of Normal Distribution?

* Because the population standard deviation is unknown, you estimate it with the sample standard deviation, which adds uncertainty.
* The t-distribution adjusts for this by having wider tails, giving more conservative (wider) confidence intervals and hypothesis tests.
* As the sample size grows, the t-distribution approaches the normal distribution.

---

### Summary:

| Distribution        | When to Use                                                      |
| ------------------- | ---------------------------------------------------------------- |
| Normal Distribution | Known population standard deviation or large samples (n > 30)    |
| T-distribution      | Unknown population standard deviation and small samples (n < 30) |


# 11. What is the Difference Between a Z-test and a T-test?

| Aspect                  | Z-test                                                                       | T-test                                                          |
| ----------------------- | ---------------------------------------------------------------------------- | --------------------------------------------------------------- |
| **Population Variance** | Known or population standard deviation ($\sigma$) known                      | Unknown; estimated using sample standard deviation (s)          |
| **Sample Size**         | Typically large (usually $n > 30$)                                           | Typically small (usually $n < 30$)                              |
| **Distribution Used**   | Standard normal distribution (Z-distribution)                                | Student’s t-distribution                                        |
| **Purpose**             | Test hypotheses about population means or proportions when variance is known | Test hypotheses about population means when variance is unknown |
| **Tail Thickness**      | Thinner tails (less variability)                                             | Heavier tails (more variability)                                |

---

### Summary:

* Use a **Z-test** when the population variance is known or the sample size is large.
* Use a **T-test** when the population variance is unknown and the sample size is small.


# 12. What is the T-test, and How Is It Used in Hypothesis Testing?

**T-test** is a statistical test used to determine if there is a significant difference between the means of one or two groups when the population variance is unknown and the sample size is small.

---

### Types of T-tests:

* **One-sample t-test:** Tests if the sample mean is different from a known or hypothesized population mean.
* **Independent two-sample t-test:** Compares the means of two independent groups.
* **Paired sample t-test:** Compares means from the same group at different times or under different conditions.

---

### How It’s Used in Hypothesis Testing:

1. **State hypotheses:**

   * Null hypothesis $H_0$: The means are equal (no difference).
   * Alternative hypothesis $H_1$: The means are different (or greater/less than).

2. **Calculate the t-statistic:**
   Based on the sample means, standard deviations, and sample sizes.

3. **Find the p-value:**
   Using the t-distribution with appropriate degrees of freedom.

4. **Make a decision:**

   * If p-value ≤ significance level ($\alpha$), reject $H_0$.
   * Otherwise, fail to reject $H_0$.

---

### Summary:

The t-test helps determine whether observed differences in sample means reflect true differences in the population or are likely due to random chance. It’s especially useful when the population variance is unknown.


# 13. What is the Relationship Between Z-test and T-test in Hypothesis Testing?

* Both **Z-test** and **T-test** are used to test hypotheses about population means.
* The **main difference** lies in whether the population standard deviation ($\sigma$) is known and the sample size.

---

### Relationship:

* When the **population standard deviation is known** and/or the **sample size is large** (usually $n > 30$), the **Z-test** is appropriate.
* When the **population standard deviation is unknown** and the **sample size is small** ($n < 30$), the **T-test** is used.
* As the sample size increases, the **T-distribution approaches the normal distribution**, so the **T-test becomes similar to the Z-test** for large samples.

---

### In short:

The **T-test** is a generalization of the **Z-test** used when extra uncertainty about the population variance exists. For large samples or known variance, both tests give nearly the same results.

# 14. What is a Confidence Interval, and How Is It Used to Interpret Statistical Results?

**Confidence Interval (CI)** is a range of values, calculated from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence (e.g., 95%).

---

### How It Works:

* A **95% confidence interval** means that if you took many samples and calculated a CI from each, about 95% of those intervals would contain the true population parameter.
* It gives an estimate of the **precision** and **uncertainty** around the sample statistic.

---

### How to Use It:

* If the confidence interval **does not include** the value stated in the null hypothesis, you can **reject the null hypothesis** at that confidence level.
* It provides more information than just a hypothesis test’s p-value by showing a range of plausible values for the population parameter.

---

### Example:

If a 95% CI for a population mean is (45, 55), it means the true mean is likely between 45 and 55 with 95% confidence.

---

### Summary:

Confidence intervals help you understand the **range of plausible values** for a population parameter and assess the **statistical significance** of your results.


# 15. What is the Margin of Error, and How Does It Affect the Confidence Interval?

**Margin of Error (MOE)** is the amount added and subtracted from a sample estimate (like a mean or proportion) to create a confidence interval. It reflects the **maximum expected difference** between the sample estimate and the true population parameter.

---

### How to Calculate Margin of Error:

$$
\text{Margin of Error} = Z_{\alpha/2} \times \text{Standard Error}
$$

Where:

* $Z_{\alpha/2}$ is the critical value from the Z-distribution (or t-distribution) based on the confidence level
* **Standard Error** measures the variability of the sample estimate

---

### Effect on Confidence Interval:

* **Confidence Interval = Sample Estimate ± Margin of Error**
* Larger margin of error → wider confidence interval → less precise estimate
* Smaller margin of error → narrower confidence interval → more precise estimate

---

### What Affects the Margin of Error?

* **Sample size**: Larger samples reduce the standard error and margin of error
* **Confidence level**: Higher confidence levels (e.g., 99% vs. 95%) increase the margin of error
* **Variability in data**: More variability increases the margin of error

---

### Summary:

The margin of error determines how wide or narrow your confidence interval is, balancing between confidence level and estimate precision.


# 16. How is Bayes' Theorem Used in Statistics, and What Is Its Significance?

**Bayes’ Theorem** is a formula that describes how to update the probability of a hypothesis based on new evidence or data.

---

### Formula:

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$

Where:

* $P(A|B)$ = Probability of hypothesis $A$ given data $B$ (posterior probability)
* $P(B|A)$ = Probability of observing data $B$ if hypothesis $A$ is true (likelihood)
* $P(A)$ = Initial probability of hypothesis $A$ before seeing data (prior probability)
* $P(B)$ = Overall probability of observing data $B$

---

### Use in Statistics:

* **Updating beliefs:** Bayes’ Theorem allows statisticians to update prior beliefs with new data to get a more accurate, “posterior” probability.
* **Bayesian inference:** It forms the foundation of Bayesian statistics, which differs from classical (frequentist) statistics by treating parameters as random variables.
* **Decision making:** Helps in areas like medical diagnosis, machine learning, and risk assessment by continuously updating probabilities with incoming data.

---

### Significance:

* Provides a **systematic way to combine prior knowledge and new evidence**.
* Enables **probabilistic reasoning** in uncertain situations.
* Supports more flexible and intuitive statistical modeling.


# 17. What is the Chi-square Distribution, and When Is It Used?

**Chi-square distribution** is a probability distribution that is used primarily in hypothesis testing and confidence interval estimation for variance and categorical data.

---

### Key Features:

* It is **right-skewed** (asymmetric) and depends on **degrees of freedom (df)**.
* The shape changes with different degrees of freedom; it approaches a normal distribution as df increases.

---

### When Is It Used?

1. **Goodness-of-fit test:**
   To check if observed categorical data fits an expected distribution.

2. **Test of independence:**
   To test whether two categorical variables are independent (e.g., in a contingency table).

3. **Test for variance:**
   To test hypotheses about the variance of a normally distributed population.

---

### Example Applications:

* Checking if dice are fair (goodness-of-fit).
* Testing if smoking status is independent of lung disease (test of independence).

---

### Summary:

The Chi-square distribution is mainly used for tests involving categorical data or population variance, helping determine whether observed data differ from expectations.


# 18. What is the Chi-square Goodness of Fit Test, and How Is It Applied?

**Chi-square Goodness of Fit Test** is a statistical test used to determine whether observed categorical data matches an expected distribution.

---

### Purpose:

* To check if the frequencies observed in categories differ significantly from the frequencies expected under a specific hypothesis.

---

### How It Is Applied:

1. **State hypotheses:**

   * Null hypothesis ($H_0$): The observed data fits the expected distribution.
   * Alternative hypothesis ($H_1$): The observed data does not fit the expected distribution.

2. **Calculate expected frequencies:**
   Based on the hypothesized distribution.

3. **Compute the test statistic:**

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where:

* $O_i$ = observed frequency in category $i$
* $E_i$ = expected frequency in category $i$

4. **Determine degrees of freedom:**
   Usually, $df = \text{number of categories} - 1$.

5. **Compare $\chi^2$ statistic to critical value** from the Chi-square distribution table or find the p-value.

6. **Make a decision:**

   * If $\chi^2$ > critical value or p-value < significance level, reject $H_0$.
   * Otherwise, fail to reject $H_0$.

---

### Example:

Testing if a six-sided die is fair by comparing observed roll counts to expected equal frequencies.

---

### Summary:

The Chi-square goodness of fit test checks whether observed categorical data align with an expected pattern or distribution.


# 19. What is the F-distribution, and When Is It Used in Hypothesis Testing?

**F-distribution** is a continuous probability distribution that arises frequently in the analysis of variance (ANOVA) and regression analysis.

---

### Key Features:

* It is **right-skewed** and depends on two sets of degrees of freedom:

  * $df_1$ (numerator degrees of freedom)
  * $df_2$ (denominator degrees of freedom)

* It is used to compare variances between groups.

---

### When Is It Used?

1. **ANOVA (Analysis of Variance):**
   To test if there are significant differences between the means of three or more groups by comparing their variances.

2. **Test for equality of variances:**
   To check if two population variances are equal.

3. **Regression analysis:**
   To test the overall significance of a regression model.

---

### Summary:

The F-distribution is used when comparing variances, especially in ANOVA and regression, to test whether group means or model fits differ significantly.


# 20. What is an ANOVA Test, and What Are Its Assumptions?

**ANOVA (Analysis of Variance)** is a statistical method used to compare the means of **three or more groups** to see if at least one group mean is significantly different from the others.

---

### Purpose:

* To test whether there are any statistically significant differences between the means of multiple groups.

---

### How It Works:

* It compares **between-group variability** to **within-group variability** using the **F-statistic**.
* A large F-value suggests that at least one group mean differs significantly.

---

### Assumptions of ANOVA:

1. **Independence:**
   Observations are independent of each other.

2. **Normality:**
   The data in each group are approximately normally distributed.

3. **Homogeneity of variances (Homoscedasticity):**
   The variances across the groups are roughly equal.

---

### Summary:

ANOVA tests if multiple group means are equal by comparing variances, assuming independence, normality, and equal variances among groups. If assumptions hold and the test is significant, you conclude that not all means are the same.


# 21. What Are the Different Types of ANOVA Tests?

There are several types of ANOVA tests depending on the study design and number of factors being analyzed:

---

### 1. **One-Way ANOVA**

* Tests for differences in means across **one independent factor** with three or more levels (groups).
* Example: Comparing test scores across three different teaching methods.

---

### 2. **Two-Way ANOVA**

* Tests for the effect of **two independent factors** on the dependent variable.
* Can assess **main effects** of each factor and the **interaction effect** between factors.
* Example: Studying the effect of teaching method and gender on test scores.

---

### 3. **Repeated Measures ANOVA**

* Used when the **same subjects** are measured multiple times under different conditions or time points.
* Controls for within-subject variability.
* Example: Measuring blood pressure of patients before, during, and after treatment.

---

### 4. **MANOVA (Multivariate ANOVA)**

* Extends ANOVA when there are **multiple dependent variables** being tested simultaneously.
* Examines the effect of factors on a vector of outcomes.

---

### Summary:

| Type                    | Number of Factors | Repeated Measures? | Purpose                                            |
| ----------------------- | ----------------- | ------------------ | -------------------------------------------------- |
| One-Way ANOVA           | 1                 | No                 | Compare means across groups                        |
| Two-Way ANOVA           | 2                 | No                 | Study effects of two factors and their interaction |
| Repeated Measures ANOVA | 1 or more         | Yes                | Analyze repeated measurements on same subjects     |
| MANOVA                  | Multiple          | No                 | Test multiple dependent variables                  |



# 22. What is the F-test, and How Does It Relate to Hypothesis Testing?

**F-test** is a statistical test that uses the **F-distribution** to compare variances and assess whether group means are significantly different.

---

### Purpose:

* To test hypotheses about the equality of **two or more group variances** or means.
* Most commonly used in **ANOVA** to test if multiple group means are equal.

---

### How It Works:

* The F-test compares the ratio of **between-group variance** to **within-group variance**.

$$
F = \frac{\text{Variance Between Groups}}{\text{Variance Within Groups}}
$$

* A **large F value** suggests that the group means are different enough to reject the null hypothesis.

---

### Role in Hypothesis Testing:

* **Null hypothesis ($H_0$)**: All group means are equal (no effect).
* **Alternative hypothesis ($H_1$)**: At least one group mean is different.
* Calculate the F-statistic from sample data and compare it to a critical value from the F-distribution or use a p-value.
* Reject $H_0$ if the F-statistic is larger than the critical value or if p-value is less than the significance level.

---

### Summary:

The F-test helps determine if group differences are statistically significant by comparing variances, forming the basis for tests like ANOVA in hypothesis testing.

```python
#2 . Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution range (e.g., integers from 1 to 10)
low, high = 1, 10  # inclusive

# Create a discrete uniform distribution object
dist = randint(low, high + 1)  # scipy randint is low (inclusive), high (exclusive)

# Generate the values (support) of the distribution
x = np.arange(low, high + 1)

# Calculate the PMF for each value
pmf = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(axis='y')
plt.show()

```

```python
#3 . Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1) - the value at which to calculate the PMF
    p : float - probability of success (0 < p < 1)

    Returns:
    float - probability P(X = x)
    """
    if x not in [0, 1]:
        return 0  # Bernoulli is only defined for 0 or 1
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7
print("P(X=1):", bernoulli_pmf(1, p))
print("P(X=0):", bernoulli_pmf(0, p))
```

```python
#4 . Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of samples to generate

# Simulate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor='black', density=True)
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(axis='y')
plt.show()
```

```python
'''1 . Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.'''

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical Z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {population_mean}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"Critical Z-value (±) = {z_critical:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Interpretation
    if abs(z_stat) > z_critical:
        print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between sample mean and population mean.")

# Example data
sample_data = np.array([102, 99, 101, 105, 98, 100, 103, 97, 104, 99])
pop_mean = 100
pop_std = 5  # known population standard deviation
alpha_level = 0.05

# Run Z-test
z_test(sample_data, pop_mean, pop_std, alpha_level)
```

```python
#3 . Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical z value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Z-Statistic = {z_stat:.4f}")
    print(f"P-Value = {p_value:.4f}")
    print(f"Critical Z-Value (±) = {z_critical:.4f}")

    # Interpretation
    if abs(z_stat) > z_critical:
        print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between sample mean and population mean.")

# Example usage
np.random.seed(0)
sample_data = np.random.normal(105, 10, size=40)  # sample data with mean ~105
population_mean = 100
population_std = 15  # known population standard deviation

one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#4 . Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data and population parameters
np.random.seed(1)
sample = np.random.normal(102, 8, size=30)  # sample with mean ~102
pop_mean = 100
pop_std = 10
alpha = 0.05

# Sample statistics
n = len(sample)
sample_mean = np.mean(sample)

# Calculate Z statistic
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# Critical Z values for two-tailed test
z_critical_low = norm.ppf(alpha / 2)
z_critical_high = norm.ppf(1 - alpha / 2)

# P-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical values: {z_critical_low:.4f}, {z_critical_high:.4f}")

# Plotting the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, 0, y, where=(x <= z_critical_low), color='red', alpha=0.4, label='Rejection Region')
plt.fill_between(x, 0, y, where=(x >= z_critical_high), color='red', alpha=0.4)

# Mark critical values
plt.axvline(z_critical_low, color='red', linestyle='--')
plt.axvline(z_critical_high, color='red', linestyle='--')

# Mark observed Z statistic
plt.axvline(z_stat, color='blue', linestyle='-', label=f'Z-statistic = {z_stat:.2f}')

plt.title('Two-tailed Z-test Decision Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5 .Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    """
    mu0: Mean under null hypothesis H0
    mu1: Mean under alternative hypothesis H1
    sigma: Standard deviation (assumed same under H0 and H1)
    alpha: Significance level (Type I error rate)
    """
    # Calculate critical value for right-tailed test
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * sigma

    # Define x range for plotting
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    # PDF under H0 and H1
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀ Distribution (μ₀)', color='blue')
    plt.plot(x, y1, label='H₁ Distribution (μ₁)', color='green')

    # Shade Type I Error region (alpha) under H0
    plt.fill_between(x, 0, y0, where=(x > x_critical), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II Error region (beta) under H1
    plt.fill_between(x, 0, y1, where=(x <= x_critical), color='orange', alpha=0.3, label='Type II Error (β)')

    # Critical value line
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value = {x_critical:.2f}')

    # Labels and legend
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05)
```

```python
#6 . Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# Generate example data for two independent samples
np.random.seed(0)
group1 = np.random.normal(loc=100, scale=10, size=30)  # Sample 1: mean ~100
group2 = np.random.normal(loc=105, scale=10, size=30)  # Sample 2: mean ~105

# Perform independent two-sample t-test
t_stat, p_value = ttest_ind(group1, group2)

# Print results
print("Independent T-Test Results")
print("--------------------------")
print(f"Sample 1 Mean: {np.mean(group1):.2f}")
print(f"Sample 2 Mean: {np.mean(group2):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("The means of the two groups are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("No significant difference between the group means.")
```

```python
#7 Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Simulate paired data (e.g., pre-test and post-test scores)
np.random.seed(1)
before = np.random.normal(loc=70, scale=5, size=30)
after = before + np.random.normal(loc=2, scale=3, size=30)  # small improvement

# Perform paired t-test
t_stat, p_value = ttest_rel(before, after)

# Print results
print("Paired Sample T-Test Results")
print("----------------------------")
print(f"Mean Before: {np.mean(before):.2f}")
print(f"Mean After: {np.mean(after):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("There is a significant difference between the paired samples.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("No significant difference between the paired samples.")

# Visualization
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(before)+1), before, 'o-', label='Before', color='blue')
plt.plot(range(1, len(after)+1), after, 'o-', label='After', color='green')
plt.title("Paired Sample Comparison")
plt.xlabel("Subject Index")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#8 . Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import norm, t, ttest_1samp

# Simulate sample data
np.random.seed(42)
population_mean = 50
population_std = 10  # Known population std for Z-test
sample = np.random.normal(loc=52, scale=10, size=25)

# Sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# ----- Z-Test -----
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

# ----- T-Test -----
t_stat, p_value_t = ttest_1samp(sample, population_mean)

# Print results
print("Z-Test vs. T-Test Comparison")
print("----------------------------")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Std Dev: {sample_std:.2f}")
print(f"Sample Size: {n}")
print()
print(f"Z-Test: Z = {z_stat:.4f}, P-value = {p_value_z:.4f}")
print(f"T-Test: T = {t_stat:.4f}, P-value = {p_value_t:.4f}")

# Interpretation
alpha = 0.05
if p_value_z < alpha:
    print("\nZ-Test Conclusion: Reject the null hypothesis.")
else:
    print("\nZ-Test Conclusion: Fail to reject the null hypothesis.")

if p_value_t < alpha:
    print("T-Test Conclusion: Reject the null hypothesis.")
else:
    print("T-Test Conclusion: Fail to reject the null hypothesis.")
```

```python
#9 . Write a Python function to calculate the confidence interval for a sample mean and explain its significance.


import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default 0.95 for 95%)

    Returns:
        tuple: (mean, lower bound, upper bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return mean, lower, upper

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=30)
confidence_interval(sample_data)
```

```python
#10 . Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        float: Margin of Error
    """
    n = len(data)
    std_dev = np.std(data, ddof=1)
    std_err = std_dev / np.sqrt(n)  # Standard error
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)  # t critical value

    moe = t_critical * std_err
    print(f"Sample Size: {n}")
    print(f"Sample Standard Deviation: {std_dev:.4f}")
    print(f"{int(confidence * 100)}% Margin of Error: ±{moe:.4f}")

    return moe

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=50, scale=8, size=25)
margin_of_error(sample_data, confidence=0.95)
```

```python
#11 . Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    """
    Apply Bayes' Theorem to calculate P(A|B)

    Parameters:
        prior_A (float): P(A) - prior probability of event A (e.g., disease)
        prob_B_given_A (float): P(B|A) - likelihood of positive test if A is true
        prob_B_given_not_A (float): P(B|¬A) - probability of positive test if A is false

    Returns:
        float: Posterior probability P(A|B)
    """
    # P(¬A)
    prior_not_A = 1 - prior_A

    # Total probability of B
    prob_B = prob_B_given_A * prior_A + prob_B_given_not_A * prior_not_A

    # Bayes' Theorem
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

# Example values
prior_disease = 0.01             # P(Disease)
sensitivity = 0.99               # P(Positive | Disease)
false_positive_rate = 0.05       # P(Positive | No Disease)

posterior = bayes_theorem(prior_disease, sensitivity, false_positive_rate)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
#12 . Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
import scipy.stats as stats

# Create a contingency table
# Rows: Gender (Male, Female)
# Columns: Product Preference (Like, Dislike)
data = [[30, 10],  # Males
        [20, 20]]  # Females

# Convert to DataFrame
table = pd.DataFrame(data, columns=["Like", "Dislike"], index=["Male", "Female"])

# Perform Chi-square test of independence
chi2, p, dof, expected = stats.chi2_contingency(table)

# Print results
print("Chi-square Test for Independence")
print("--------------------------------")
print("Observed Table:")
print(table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=["Like", "Dislike"], index=["Male", "Female"]))
print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Interpretation
alpha = 0.05
if p < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("There is a significant association between gender and product preference.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("No significant association between gender and product preference.")
```

```python
#13 .Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data.


import pandas as pd
from scipy.stats import chi2_contingency

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test.

    Parameters:
        observed_data (list of lists or 2D array): Contingency table

    Returns:
        DataFrame: Expected frequency table
    """
    # Convert to DataFrame for readability
    df_observed = pd.DataFrame(observed_data)

    # Calculate expected frequencies using chi2_contingency
    _, _, _, expected = chi2_contingency(df_observed)

    df_expected = pd.DataFrame(expected, index=df_observed.index, columns=df_observed.columns)

    print("Observed Frequencies:")
    print(df_observed)
    print("\nExpected Frequencies:")
    print(df_expected.round(2))

    return df_expected

# Example observed data (2x2 table)
observed = pd.DataFrame(
    [[25, 15],
     [20, 40]],
    index=["Group A", "Group B"],
    columns=["Success", "Failure"]
)

# Calculate expected frequencies
calculate_expected_frequencies(observed)
```
